```python
import pandas as pd                                   
from bs4 import BeautifulSoup                          
import requests  
import datetime
from time import sleep
```

```python
headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.86 Safari/537.36'
}
```

```python
data = []
list = {}
counter = 0
for i in range(1,34):
    counter += 1
    print(counter)
    url = f'https://www.realtor.com/realestateagents/94040/photo-1/pg-{i}'
    r = requests.get(url, headers = headers)
    soup = BeautifulSoup(r.content, 'html5lib')
    # print(r.content)



    name_list1 = soup.findAll('div',attrs={'class':'jsx-3970352998 agent-name text-bold'})
    name_list2 = []
    for i in name_list1:
        n = i.get_text()
        name_list2.append(n)

    name_list = []
    for i in name_list2:
           if i not in name_list:
                name_list.append(i)



    phone_num1 = soup.findAll('div',attrs={'class':'jsx-3970352998 agent-phone hidden-xs hidden-xxs'})
    phone_num2 = []
    for i in phone_num1:
        n = i.get_text()
        phone_num2.append(n)
    # phone_num = list(OrderedDict.fromkeys(phone_num2))
    phone_num = []
    for i in phone_num2:
           if i not in phone_num:
                phone_num.append(i)
#     print(phone_num)

    sleep(3)
    for name,phone_number in zip(name_list,phone_num):
        name1 = name.split(' ')
        name2 = name
        first_name = name1[0]
        last_name = name2.replace(f'{first_name}',' ').replace('Team',' ')
    #     print(first_name)
    #     print(last_name)
    #     print(phone_number)
        list = {
            'First Name': first_name,
            'Last Name': last_name,
            'Phone Number': phone_number
        }
        data.append(list)
    

df = pd.DataFrame(data)
df.to_excel('realtor_leads_94040.xlsx',encoding='utf-8-sig', index=False)
    
```
